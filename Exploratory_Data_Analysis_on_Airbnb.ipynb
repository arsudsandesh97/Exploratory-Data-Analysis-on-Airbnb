## changing data types

```python
# changing data types

# Room Type
df['room_type'] = df['room_type'].astype('category')

# Last Review Date
df['last_review'] = df['last_review'].astype('datetime64[ns]')
```

## Handling missing values

```python
# Handling misssing values

# Name
df['name'] = df['name'].fillna('Unknown')

# Host Name
df['host_name'] = df['host_name'].fillna('Unknown')

# Last Review
df['last_review'] = df['last_review'].fillna(df['last_review'].mode()[0])

# Reviews per month
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean())

# Printing the count of null values
df.isnull().sum()
```

## Keeping only required columns

```python
# Keeping required columns
required_columns = ['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type','price','minimum_nights','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count','availability_365']
df = df[required_columns]

# printing updated df
df
```

## Outlier Detection

```python
# Visualizing outliers using box plot for price column.
sns.boxplot(x=df['price'])
```

```python
# Outlier Detection Using IQR (Inter Quartile Range)
def outlier_detection_using_IQR(DF_column):
  Q1 = np.percentile(DF_column,25)
  Q3 = np.percentile(DF_column,75)
  IQR = Q3-Q1
  lower_limit = Q1 - (1.5*IQR)
  upper_limit = Q3 + (1.5*IQR)
  return lower_limit,upper_limit
```

```python
lower_bound,upper_bound = outlier_detection_using_IQR(df['price'])

df = df[(df.price>lower_bound) & (df.price<upper_bound)]
```

```python
# Visualizing outliers using box plot for updated price column.
sns.boxplot(x=df['price'])
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
df
```

```python
# Total No of listings By location

location_count = df['neighbourhood_group'].value_counts()

top_locations = location_count.reset_index()

top_locations.columns = ['neighbourhood_group','listings_count']

top_locations
```

```python
# Chart - 1 visualization code

sns.barplot(x=top_locations['neighbourhood_group'],
            y=top_locations['listings_count'],
            palette='husl')
plt.title('Total No of Listings By Location')
plt.xlabel('Location')
plt.ylabel('No of Listings')
plt.show()
```

#### Chart - 2

```python
# Price distribution by location

price_distribution = df.groupby('neighbourhood_group')['price'].sum().sort_values(ascending=False)

price_distribution = price_distribution.reset_index()

price_distribution.columns = ['Location','Total_Price']

price_distribution
```

```python
sns.barplot(x=price_distribution['Location'],
            y=price_distribution['Total_Price'],
            palette='husl')
plt.show()
```

#### Chart - 3

```python
# Top N Hosts by listing count
top_5_hosts = df.groupby('host_name')['calculated_host_listings_count'].sum().sort_values(ascending=False).head(5)

top_5_hosts = top_5_hosts.reset_index()

top_5_hosts.columns = ['host_name','listings_count']

top_5_hosts
```

```python
# Chart - 3 visualization code

sns.barplot(x=top_5_hosts['host_name'],
            y=top_5_hosts['listings_count'],
            palette='husl')
plt.title('Top 5 Hosts by Listing Count')
plt.xlabel('Host Name')
plt.ylabel('No of Listings')
plt.show()
```

#### Chart - 4

```python
# Listings By Room Type

listings_by_room_type = df['room_type'].value_counts()

listings_by_room_type = listings_by_room_type.reset_index()

listings_by_room_type.columns = ['room_type','listings_count']

listings_by_room_type
```

#### Chart - 6

```python
# Total No of Reviews by Room Type

no_of_reviews_by_room_type = df.groupby('room_type')['number_of_reviews'].sum().sort_values(ascending = False)

no_of_reviews_by_room_type = no_of_reviews_by_room_type.reset_index()

no_of_reviews_by_room_type.columns = ['Room Type','Total No of Reviews']

no_of_reviews_by_room_type
```
